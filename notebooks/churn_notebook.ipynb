**Columns and what they are mean**:

**Personal information**:
- **PERSON_ID**: Unique identifier for the individual.
- **AGE**: The age of the person in years.
- **STATE**: The person's membership or program status (e.g., FROZEN, ACTIVE).
- **PERSON_TYPE**: The type or classification of the person (e.g., PRIVATE, CORPORATE).
- **GENDER**: The person’s gender (e.g., M for Male, F for Female).
- **CENTER_REGION**: The region of the center associated with the person (e.g., WR for Western Region).
- **CITY**: The city where the center or the person is located.
- **TOT_SUBS**: Total number of subscriptions the person has.
- **REJOIN_CNT**: The count of times the person rejoined after cancellation.
- **RENEWAL_CNT**: The count of subscription renewals by the person.
- **NEWSALE_CNT**: The count of new sales made to the person.
- **NO_OF_CENTER**: The total number of centers associated with the person.
- **SUBS_CREATION_DATE**: The date when the person's subscription was created.
- **OUTAGE**: Indicates if the service had any outages.
- **SUBS_DAYS**: Total number of days the subscription has been active.
- **OUTAGE_PERC**: The percentage of time the subscription was affected by outages.
- **OUTAGE_TILLNOW**: The cumulative time of service outages until now.

**Programs names**:
- **GX**: program name.
- **PT**: program name.
- **BOTH_PROGRAM**: Indicates if the person is enrolled in multiple programs.
- **FT90_P**: program name.
- **NO_PRGRAM**: Indicates that the person is not currently enrolled in any programs.

**Type of Center**
- **FT_CENTER**: Indicates involvement with a full-time center.
- **PRO_CENTER**: Indicates involvement with a professional center.
- **PLUS_CENTER**: Indicates involvement with a plus center.
- **XPRESS_CENTER**: Indicates involvement with an express center.
- **POPUP_CENTER**: Indicates involvement with a pop-up center.
- **JUNIOR_CENTER**: Indicates involvement with a junior center.
- **HQ_CENTER**: Indicates involvement with the headquarters center.
- **NO_OF_PRODUCTS**: The number of products the person has purchased or subscribed to.

**PKGs**:
- **MONTH_12_PKG**: The status of a 12-month subscription package.
- **MONTH_9_PKG**: The status of a 9-month subscription package.
- **MONTH_6_PKG**: The status of a 6-month subscription package.
- **MONTH_3_PKG**: The status of a 3-month subscription package.
- **MONTH_1_PKG**: The status of a 1-month subscription package.
- **DAYS_1_PKG**: The status of a 1-day subscription package.

**Dates**:
- **Visits : YYYY / MM**: The number of visits made by the person during a specific year and month.
- **Subscribe : YYYY / MM**: The status of the person’s subscription during a specific year and month.

```python
# Function to extract first and final value and check for multiple unique values
def process_center_region(df):
    # Extract first and final value
    df['CENTER_REGION_FIRST'] = df['CENTER_REGION'].apply(lambda x: x.split('-->')[0])  # First value
    df['CENTER_REGION_LAST'] = df['CENTER_REGION'].apply(lambda x: x.split('-->')[-1])  # Last value
    
    # Check if there are multiple unique values
    df['CENTER_REGION_CHANGE'] = df['CENTER_REGION'].apply(lambda x: 'True' if len(set(x.split('-->'))) > 1 else 'False')

    return df
```

```python
# from the analysis:
# I will keep PRIVATE and CORPORATE only cause the model will not lean other values cause they are have very low numbers!
# Then PRIVATE -> 0 , CORPORATE->1
# GENDER M -> 0, F -> 1
# GX No Showup -> 0, GX Showup -> 1
# PT No Showup -> 0, PT Showup -> 1
# FT90_P No Showup->0, FT90 Showup -> 1
# NO_PRGRAM Go’s Solo->0, At least one program->1
# CENTER_REGION_CHANGE False ->0, True->1
# CITY_CHANGE False ->0, True->1

# BOTH_PROGRAM -> drop cause we have data for each program!

# CITY_FIRST -> we have 35 value! Target Encoding -> " mean of the target variable"

# CENTER_REGION_FIRST
# CR     916697
# WR     644324
# ER     245570
# AJ      17271
# SH      12417
# DU      10524
# RAK      9732

###
# Target : STATE_FIRST -> churn 
# INACTIVE -> 0 -> churn
# ACTIVE -> 1 -> not churn
```

```python
# Drop all rows with NaN values from the DataFrame
df_gold.dropna(inplace=True)

# 1. PERSON_TYPE: Keep only 'PRIVATE' and 'CORPORATE', then map PRIVATE -> 0, CORPORATE -> 1
df_gold = df_gold[df_gold['PERSON_TYPE'].isin(['PRIVATE', 'CORPORATE'])]
df_gold['PERSON_TYPE'] = df_gold['PERSON_TYPE'].str.strip()  # Remove any extra spaces
df_gold['PERSON_TYPE'] = df_gold['PERSON_TYPE'].map({'PRIVATE': 0, 'CORPORATE': 1})

# 2. GENDER: M -> 0, F -> 1
df_gold['GENDER'] = df_gold['GENDER'].map({'M': 0, 'F': 1})

# 3. GX: Clean values and map No Showup -> 0, GX Showup -> 1
df_gold['GX'] = df_gold['GX'].str.strip()  # Remove any extra spaces
df_gold['GX'] = df_gold['GX'].map({'No Showup': 0, 'GX Showup': 1})

# 4. PT: Clean values and map No Showup -> 0, PT Showup -> 1
df_gold['PT'] = df_gold['PT'].str.strip()  # Remove any extra spaces
df_gold['PT'] = df_gold['PT'].map({'No Showup': 0, 'PT Showup': 1})

# 5. FT90_P: Clean values and map No Showup -> 0, FT90 Showup -> 1
df_gold['FT90_P'] = df_gold['FT90_P'].str.strip()  # Remove any extra spaces
df_gold['FT90_P'] = df_gold['FT90_P'].map({'No Showup': 0, 'FT90 Showup': 1})

# 6. NO_PRGRAM: Clean values and map Go’s Solo -> 0, At least one program -> 1
df_gold['NO_PRGRAM'] = df_gold['NO_PRGRAM'].str.strip()  # Remove any extra spaces
df_gold['NO_PRGRAM'] = df_gold['NO_PRGRAM'].map({'Go’s Solo': 0, 'At least one program': 1})

# 7. CENTER_REGION_CHANGE: Clean values and map False -> 0, True -> 1
df_gold['CENTER_REGION_CHANGE'] = df_gold['CENTER_REGION_CHANGE'].str.strip()  # Remove any extra spaces
df_gold['CENTER_REGION_CHANGE'] = df_gold['CENTER_REGION_CHANGE'].map({'False': 0, 'True': 1})

# 8. CITY_CHANGE: Clean values and map False -> 0, True -> 1
df_gold['CITY_CHANGE'] = df_gold['CITY_CHANGE'].str.strip()  # Remove any extra spaces
df_gold['CITY_CHANGE'] = df_gold['CITY_CHANGE'].map({'False': 0, 'True': 1})

# 9. BOTH_PROGRAM: Drop column
df_gold.drop('BOTH_PROGRAM', axis=1, inplace=True)

import joblib

# 10. CITY_FIRST: Target Encoding (Mean of the target variable 'STATE_FIRST' where 0 = churn, 1 = not churn)
df_gold['STATE_FIRST'] = df_gold['STATE_FIRST'].map({'INACTIVE': 0, 'ACTIVE': 1})
city_means = df_gold.groupby('CITY_FIRST')['STATE_FIRST'].mean()
df_gold['CITY'] = df_gold['CITY_FIRST'].map(city_means)
df_gold.drop(columns=["CITY_FIRST"], axis=1, inplace=True)

# Save city_means for future use
joblib.dump(city_means, r'\models\city_encoding.pkl')

# 11. CENTER_REGION_FIRST: Target encoding
region_means = df_gold.groupby('CENTER_REGION_FIRST')['STATE_FIRST'].mean()
df_gold['CENTER_REGION'] = df_gold['CENTER_REGION_FIRST'].map(region_means)
df_gold.drop('CENTER_REGION_FIRST', axis=1, inplace=True)

# Save region_means for future use
joblib.dump(region_means, r'\models\region_encoding.pkl')

# 12. Rename STATE_FIRST to CHURN and map INACTIVE -> 0, ACTIVE -> 1
df_gold.rename(columns={'STATE_FIRST': 'CHURN'}, inplace=True)
df_gold['CHURN'] = df_gold['CHURN'].map({0: 0, 1: 1})  # 0 -> churn, 1 -> not churn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
```

```python
# Initialize the models
rf = RandomForestClassifier(random_state=42)
# Train the models
rf.fit(X_train_scaled, y_train)
# Predict the test set
y_pred_rf = rf.predict(X_test_scaled)
```

```python
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train_scaled, y_train)
y_pred_xgb = xgb.predict(X_test_scaled)
```

```python
ada = AdaBoostClassifier(random_state=42)
ada.fit(X_train_scaled, y_train)
y_pred_ada = ada.predict(X_test_scaled)
```

```python
# Plot confusion matrix for each model
models = ['Random Forest', 'XGBoost', 'AdaBoost']
predictions = [y_pred_rf, y_pred_xgb, y_pred_ada]
classifiers = [rf, xgb, ada]

for i, (model, y_pred) in enumerate(zip(models, predictions)):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

# Initialize models and predictions (assuming these are already defined)
models = ['Random Forest', 'XGBoost', 'AdaBoost']
predictions = [y_pred_rf, y_pred_xgb, y_pred_ada]

# Print confusion matrix, accuracy, and F1 score for each model
for model, y_pred in zip(models, predictions):
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f'{model} Confusion Matrix:')
    print(cm)
    print(f'Accuracy: {accuracy:.4f}')
    print(f'F1 Score: {f1:.4f}')
    print('-' * 30)
```

```python
# Feature importance for each

import matplotlib.pyplot as plt
import numpy as np

# Create a function to plot feature importance
def plot_feature_importance(importance, features, model_name):
    indices = np.argsort(importance)[::-1]  # Sort feature importance in descending order
    plt.figure(figsize=(10, 6))
    plt.title(f"{model_name} Feature Importance")
    plt.bar(range(X_train_scaled.shape[1]), importance[indices], align='center')
    plt.xticks(range(X_train_scaled.shape[1]), np.array(features)[indices], rotation=90)
    plt.tight_layout()
    plt.show()

# Get the feature importance for Random Forest
rf_importance = rf.feature_importances_
plot_feature_importance(rf_importance, X_train.columns, "Random Forest")

# Get the feature importance for AdaBoost
ada_importance = ada.feature_importances_
plot_feature_importance(ada_importance, X_train.columns, "AdaBoost")

# Get the feature importance for XGBoost
xgb_importance = xgb.feature_importances_
plot_feature_importance(xgb_importance, X_train.columns, "XGBoost")
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
import joblib  # for saving models

# Create weighted ensemble
ensemble_model = VotingClassifier(
    estimators=[('xgb', best_xgb), ('rf', rf)],  
    voting='soft',  # Use soft voting to allow probability weighting
    weights=[0.55, 0.45]  # Initial weights, adjust based on performance metrics
)

# Fit the ensemble model
ensemble_model.fit(X_train_scaled, y_train)
```

```python
# Predictions on test set
y_pred_ensemble = ensemble_model.predict(X_test_scaled)

# Print Confusion Matrix
cm = confusion_matrix(y_test, y_pred_ensemble)
print("Confusion Matrix:\n", cm)

# Print F1 Score
f1 = f1_score(y_test, y_pred_ensemble)
print("F1 Score:", f1)

# Print Accuracy
accuracy = accuracy_score(y_test, y_pred_ensemble)
print("Accuracy:", accuracy)
```

```python
# Save the ensemble model
joblib.dump(ensemble_model, 'ensemble_model.pkl')
```
